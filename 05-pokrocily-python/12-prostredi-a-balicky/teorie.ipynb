# Správa balíčků v Pythonu

V Jupyter Notebooku (zvláště v DevContaineru) už běžíme v nějakém prostředí. Můžeme ale používat `pip` přímo z buňky pro instalaci knihoven do tohoto prostředí.

**Pozor:** V reálném projektu v terminálu byste nejprve aktivovali virtuální prostředí (`source .venv/bin/activate`), v notebooku to za nás často řeší kernel.

## 1. Zjištění, kde se nacházíme
Podíváme se, jaký Python kernel používá a kde jsou nainstalované balíčky.

```python
import sys
print(f"Cesta k Pythonu: {sys.executable}")
print(f"Cesta k balíčkům: {sys.path[-1]}")
```

## 2. Instalace balíčku (`pip install`)
Zkusíme nainstalovat knihovnu `colorama`, pokud ji ještě nemáme. Použijeme magický příkaz `%pip`, který zaručuje instalaci do *aktuálního* kernelu (na rozdíl od `!pip`, který může instalovat jinam).

```python
# Instalace knihovny colorama
%pip install colorama
```

## 3. Použití nainstalovaného balíčku
Teď můžeme knihovnu importovat.

```python
from colorama import Fore, Style

print(Fore.BLUE + "Tento text by měl být modrý!" + Style.RESET_ALL)
print(Fore.YELLOW + "A tento žlutý!" + Style.RESET_ALL)
```

## 4. Výpis závislostí (`pip freeze`)
Podíváme se, co všechno je v prostředí nainstalované.

```python
%pip freeze | grep colorama
```
