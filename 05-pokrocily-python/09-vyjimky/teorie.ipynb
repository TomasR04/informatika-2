# Výjimky a ošetřování chyb

Výjimka je objekt, který reprezentuje chybu. Pokud není zachycena, probublá až nahoru a ukončí program.

## 1. Co se stane bez ošetření?
Zkuste spustit následující buňku. Uvidíte `Traceback` (výpis zásobníku).

```python
# Neošetřené dělení nulou
print(10 / 0)
print("Tento text se už nevypíše, program spadl.")
```

## 2. Ošetření pomocí `try-except`
Nyní stejnou chybu zachytíme.

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Chyba: Pokus o dělení nulou!")

print("Program pokračuje dál...")
```

## 3. Práce s více typy chyb
Můžeme reagovat různě na různé problémy.

```python
def bezpecny_vstup(hodnota):
    try:
        cislo = int(hodnota)
        vysledek = 100 / cislo
        return vysledek
    except ValueError:
        print("To není číslo!")
    except ZeroDivisionError:
        print("Nulou dělit nelze!")
    except Exception as e:
        print(f"Jiná chyba: {e}")

bezpecny_vstup("abc")
bezpecny_vstup(0)
bezpecny_vstup(20)
```

## 4. `else` a `finally`
* `else`: Běží, když chyba nenastala.
* `finally`: Běží vždy (úklid).

```python
try:
    print("1. Otevírám spojení...")
    x = 10 / 2 # Zkuste změnit na 10 / 0
except ZeroDivisionError:
    print("2. Chyba!")
else:
    print("2. Vše OK.")
finally:
    print("3. Zavírám spojení (provádím vždy).")
```
