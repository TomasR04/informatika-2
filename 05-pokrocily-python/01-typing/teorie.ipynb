# Type Hinting (Typové anotace)

Python 3.5+ přinesl možnost definovat datové typy u proměnných a funkcí. To z Pythonu nedělá staticky typovaný jazyk (jako C++), ale pomáhá nám to psát bezpečnější kód.

Typy se píší za dvojtečku `:`.

```python
# Základní typy
jmeno: str = "Alice"
vek: int = 30
vyska: float = 1.75
student: bool = False

print(f"{jmeno} má {vek} let.")
```

## Typování funkcí
U funkcí definujeme typy argumentů a pomocí šipky `->` typ návratové hodnoty.

```python
def scitani(a: int, b: int) -> int:
    return a + b

def pozdrav(jmeno: str) -> None:  # None znamená, že funkce nic nevrací
    print(f"Ahoj {jmeno}!")

vysledek = scitani(10, 20)
pozdrav("Karle")

# Pozor: Toto Python spustí bez chyby, ale editor by měl podtrhnout chybu!
chyba = scitani("Ahoj", "Světe") 
print(chyba)
```

## Složené typy (Seznamy a Slovníky)
Často nepracujeme jen s čísly, ale se seznamy čísel atd. Používáme syntaxi `typ[podtyp]`.

```python
# Seznam řetězců
mesta: list[str] = ["Praha", "Brno", "Ostrava"]

# Slovník: Jméno (str) -> Věk (int)
lide: dict[str, int] = {"Petr": 25, "Jana": 22}

# Seznam seznamů čísel (Matice)
matice: list[list[int]] = [
    [1, 2, 3],
    [4, 5, 6]
]
```

## Union a Optional
Někdy proměnná může nabývat více typů.
* `int | float` znamená "číslo celé nebo desetinné".
* `str | None` znamená "buď text, nebo nic".

```python
def vydel(a: float, b: float) -> float | None:
    if b == 0:
        return None
    return a / b

vysledek1 = vydel(10, 2)
vysledek2 = vydel(5, 0)

print(vysledek1)
print(vysledek2)
```
