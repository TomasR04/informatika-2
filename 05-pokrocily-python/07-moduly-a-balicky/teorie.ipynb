# Moduly a Importování

Abychom demonstrovali práci s moduly v Jupyter Notebooku, využijeme "magický příkaz" `%%writefile`. Ten nám umožní vytvořit soubor `.py` přímo odsud.

## 1. Vytvoření modulu
Vytvoříme soubor `moje_funkce.py` s jednoduchou funkcí.

```python
%%writefile moje_funkce.py

def pozdrav(jmeno):
    return f"Ahoj {jmeno}, zdravím tě z modulu!"

PI = 3.14159

# Tento print se provede okamžitě při importu (pokud není chráněn)
print("Modul 'moje_funkce' byl načten!")
```

## 2. Importování modulu
Nyní, když soubor existuje, můžeme ho importovat.

```python
import moje_funkce

# Použití funkce z modulu
zprava = moje_funkce.pozdrav("Student")
print(zprava)

# Použití proměnné z modulu
print(f"Hodnota PI je: {moje_funkce.PI}")
```

## 3. Selektivní import (from ... import)
Nemusíme importovat celý modul, stačí jen to, co potřebujeme.

```python
from moje_funkce import pozdrav

# Nyní voláme funkci přímo, bez tečky
print(pozdrav("Karel"))
```

## 4. Ochrana `if __name__ == "__main__"`
Všimněte si, že v prvním kroku se vypsalo "Modul byl načten!". To je často nežádoucí. Ukážeme si, jak to opravit vytvořením lepšího modulu.

```python
%%writefile lepsi_modul.py

def vypocet(x):
    return x * x

# Toto se provede jen při přímém spuštění, ne při importu
if __name__ == "__main__":
    print("Provádím testy modulu...")
    print(f"Test výpočtu: {vypocet(5)}")
```

```python
import lepsi_modul

print("Modul importován. Vidíte výpis testů? Neměli byste.")
print(lepsi_modul.vypocet(10))
```
