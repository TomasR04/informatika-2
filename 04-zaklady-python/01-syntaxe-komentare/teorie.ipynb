# 01 - Syntaxe, Výstup a Komentáře

V tomto interaktivním sešitu si vyzkoušíme úplné základy Pythonu. Nebudeme jen číst, ale kód rovnou spouštět.

## 1. Funkce `print()`
Základní příkaz pro komunikaci počítače s člověkem. Vypíše text do konzole.

```python
print("Ahoj, světe!")
print(42)
print(3.14)
```

### Parametry `sep` a `end`
Standardně `print` odděluje položky mezerou a na konci udělá nový řádek. To můžeme změnit.

```python
# Parametr sep="..." (separator)
print("jan", "novak", "email.cz", sep="@")

# Parametr end="..." (co bude na konci)
print("Načítám", end="...")
print("Hotovo!")
```

## 2. F-strings (Formátované řetězce)
Moderní způsob, jak vkládat data do textu. Před uvozovky dáme `f` a proměnné píšeme do `{}`.

```python
jmeno = "Alice"
body = 150

# Použití f-stringu
print(f"Hráč {jmeno} získal {body} bodů.")

# Můžeme v nich i počítat
print(f"Za 10 kol by získal {body * 10} bodů.")
```

## 3. Odsazování (Indentation)
Toto je nejčastější chyba začátečníků. Python nepoužívá `{}`. Bloky kódu se dělají odsazením (Tab nebo 4 mezery). Kód, který patří k sobě, musí být zarovnaný.

```python
print("Začátek programu")

if True:
    print("    Tento řádek je odsazený a je součástí podmínky.")
    print("    Tento taky.")

print("Tento řádek už odsazený není a pokračuje dál.")
```
