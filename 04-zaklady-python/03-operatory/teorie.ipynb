## 1. Aritmetické operátory
Základní počty znáte (`+`, `-`, `*`). V Pythonu je ale pár specialit.

### Dělení (`/`) vs Celochíselné dělení (`//`)
Zkuste spustit následující kód a všímejte si rozdílu.

```python
print("Klasické dělení (10 / 3):", 10 / 3)
print("Celochíselné dělení (10 // 3):", 10 // 3)
print("Dělení se zbytkem (10.0 // 3):", 10.0 // 3) # I když je vstup float, výsledek je 'celé číslo' (typ float)
```

### Mocnění (`**`)
V jiných jazycích se často používá `^` (což v Pythonu znamená něco jiného - bitový XOR). Pro mocninu používejte dvě hvězdičky.

```python
print("2 na třetí (2 ** 3):", 2 ** 3)
print("Odmocnina ze 16 (16 ** 0.5):", 16 ** 0.5)
```

## 2. Modulo (`%`)
Modulo vrací **zbytek po dělení**. Je to jeden z nejužitečnějších operátorů v programování. 

**Kde se používá?**
1. Zjištění, zda je číslo sudé/liché.
2. Omezení čísla do určitého rozsahu (např. hodiny na ciferníku).
3. Cyklování barev nebo pozic.

```python
cislo = 17
delitel = 5

vysledek = cislo // delitel
zbytek = cislo % delitel

print(f"{cislo} děleno {delitel} je {vysledek} a zbytek je {zbytek}")
```

### Úkol: Sudá nebo lichá?
Zkuste změnit hodnotu v proměnné `test_cislo` a spusťte kód. Pokud je zbytek po dělení 2 nula, číslo je sudé.

```python
test_cislo = 42

zbytek_po_deleni_dvema = test_cislo % 2

print(f"Číslo {test_cislo} má po dělení dvojkou zbytek: {zbytek_po_deleni_dvema}")
print(f"Je číslo sudé? {zbytek_po_deleni_dvema == 0}")
```

## 3. Logické operátory (`and`, `or`, `not`)
Logika je srdcem rozhodování programu. 
* `and`: Musí platit VŠECHNO.
* `or`: Stačí, aby platilo ALESPOŇ JEDNO.
* `not`: Obrátí pravdu na lež a naopak.

```python
mam_penize = True
mam_cas = False
je_vikend = True

# Scénář 1: Jdu do kina, pokud mám peníze A čas.
jdu_do_kina = mam_penize and mam_cas
print(f"Jdu do kina? {jdu_do_kina}")

# Scénář 2: Jdu na výlet, pokud mám čas NEBO je víkend.
jdu_na_vylet = mam_cas or je_vikend
print(f"Jdu na výlet? {jdu_na_vylet}")

# Scénář 3: Nejdu do práce (negace)
jdu_do_prace = not je_vikend
print(f"Jdu do práce? {jdu_do_prace}")
```

## 4. Porovnávací operátory a Řetězení
Python umí něco, co jiné jazyky (jako C++ nebo Java) neumí jednoduše - řetězit porovnání. Můžeme napsat `a < b < c`.

```python
teplota = 22

# Klasický zápis (funguje všude)
je_prijemne_klasicky = (teplota >= 18) and (teplota <= 25)

# Pythonic zápis (řetězení)
je_prijemne_python = 18 <= teplota <= 25

print(f"Klasicky: {je_prijemne_klasicky}")
print(f"Pythonic: {je_prijemne_python}")
```
