# 02 - Proměnné a Datové typy

V tomto sešitu prozkoumáme, jak Python ukládá data. Narozdíl od C++ nebo Javy je Python "dynamicky typovaný", což znamená větší volnost, ale také nutnost dávat si pozor na to, co v proměnné zrovna máme.

## 1. Základní datové typy
Zkuste spustit buňku a podívejte se, jak funkce `type()` prozradí datový typ. Zahrnuli jsme zde všechny základní typy.

```python
# Ciselne typy
cislo_int = 10
cislo_float = 3.14
cislo_complex = 2 + 3j

# Textove a bytove typy
text = "Python"
data_bytes = b"Ahoj"

# Ostatni
pravda = True
nic = None

print(f"Hodnota: {cislo_int}, Typ: {type(cislo_int)}")
print(f"Hodnota: {cislo_float}, Typ: {type(cislo_float)}")
print(f"Hodnota: {cislo_complex}, Typ: {type(cislo_complex)}")
print(f"Hodnota: {text}, Typ: {type(text)}")
print(f"Hodnota: {data_bytes}, Typ: {type(data_bytes)}")
print(f"Hodnota: {pravda}, Typ: {type(pravda)}")
print(f"Hodnota: {nic}, Typ: {type(nic)}")
```

## 2. Dynamické typování
V Pythonu můžete do jedné proměnné uložit číslo a o řádek níže do té samé proměnné uložit text. Původní hodnota se přepíše.

```python
x = 100
print(f"x je nyní {x} ({type(x)})")

x = "Už nejsem číslo"
print(f"x je nyní {x} ({type(x)})")
```

## 3. Nejčastější chyba: Sčítání textu a čísel
Následující kód skončí chybou `TypeError`. Python neumí sečíst text a číslo dohromady. Spusťte ho a prohlédněte si chybovou hlášku.

```python
vek = 20
text = "Je mi "

# Toto spadne:
vysledek = text + vek
print(vysledek)
```

### Jak to opravit? (Casting)
Musíme číslo převést na text pomocí funkce `str()`. Opravte kód v buňce níže:

```python
vek = 20
text = "Je mi "

# Oprava: str(vek)
vysledek = text + str(vek)
print(vysledek)
```
