# Funkce v Pythonu
Funkce nám pomáhají organizovat kód, aby byl přehlednější a abychom nemuseli psát to samé několikrát (princip DRY - Don't Repeat Yourself).

## 1. Definice a volání funkce
Funkci definujeme klíčovým slovem `def`. Tělo funkce musí být odsazené.

```python
def pozdrav():
    print("Ahoj! Vítám tě u kurzu Pythonu.")
    print("Doufám, že se ti bude dařit.")

# Tady funkci zavoláme (spustíme)
pozdrav()
pozdrav() # Můžeme ji volat kolikrát chceme
```

## 2. Funkce s parametry
Funkci můžeme předat data pomocí parametrů v závorkách.

```python
def privitani(jmeno):
    print(f"Vítej, uživateli {jmeno}!")

privitani("Petr")
privitani("Jana")
```

## 3. Návratová hodnota (`return`)
Funkce často slouží k výpočtům. Výsledek nám vrátí pomocí `return`. Tento výsledek si můžeme uložit do proměnné.

```python
def scitani(a, b):
    vysledek = a + b
    return vysledek

součet = scitani(10, 5)
print(f"Výsledek sčítání je: {součet}")

# Můžeme volat i přímo ve funkci print
print(scitani(100, 200))
```

## 4. Výchozí (defaultní) argumenty
Parametrům můžeme nastavit výchozí hodnotu. Pokud ji uživatel při volání nezadá, použije se ta naše.

```python
def predstaveni(jmeno, pozice="Programátor"):
    print(f"Jmenuji se {jmeno} a pracuji jako {pozice}.")

predstaveni("Lukáš")             # Použije se defaultní pozice
predstaveni("Eva", "Manažerka")  # Pozice se přepíše
```

## 5. Lambda funkce (Anonymní funkce)
Krátké funkce zapsané na jeden řádek. Často se používají, když potřebujete jednoduchou funkci jen na jednom místě.

```python
# Lambda funkce, která vynásobí číslo dvěma
dvojnasobek = lambda x: x * 2

print(dvojnasobek(5))   # Vypíše 10
print(dvojnasobek(10))  # Vypíše 20
```
