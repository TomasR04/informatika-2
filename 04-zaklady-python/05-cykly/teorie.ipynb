# 05 - Cykly (Loops)

Cykly jsou jedním ze základních pilířů algoritmizace. Umožňují počítači dělat to, co mu jde nejlépe: opakovat úkoly rychle a bez chyb.

## 1. Cyklus `for` a funkce `range()`

Funkce `range()` generuje sekvenci čísel. Je důležité si pamatovat, že horní mez (parametr `stop`) se do sekvence **nepočítá**.

````python
# range(stop) - od 0 do stop-1
print("range(5): ", end="")
for i in range(5):
    print(i, end=" ")
print() # Odradkovani

# range(start, stop) - od start do stop-1
print("range(2, 6): ", end="")
for i in range(2, 6):
    print(i, end=" ")
print()

# range(start, stop, step) - skace po kroku
print("range(0, 10, 2): ", end="")
for i in range(0, 10, 2):
    print(i, end=" ")
````

## 2. Cyklus `while`

`while` cyklus se používá, když podmínka ukončení závisí na něčem jiném než na počtu opakování (např. na vstupu uživatele nebo stavu aplikace).

````python
pocet_zivotu = 3

while pocet_zivotu > 0:
    print(f"Hrdina bojuje! Zbyva zivotu: {pocet_zivotu}")
    pocet_zivotu -= 1 # Odebereme zivot
    
print("Hrdina padl.")
````

## 3. Příkazy `break` a `continue`

- **`break`**: Okamžitě opustí cyklus.
- **`continue`**: Ukončí aktuální průběh a začne znovu od začátku cyklu (s další hodnotou).

````python
print("Hledam cislo delitelne 7 v rozsahu 1-50:")

for i in range(1, 51):
    if i % 7 != 0:
        continue # Neni to ono, jdeme na dalsi cislo
    
    print(f"Nalezeno prvni cislo: {i}")
    break # Nasli jsme prvni, koncime, nemusime hledat dal
````

## 4. Vnořené cykly (Nested Loops)

Jeden cyklus může být uvnitř druhého. To je typické pro práci s mřížkou (souřadnice x, y) nebo tabulkami.
V následujícím příkladu vykreslíme malou násobilku.

````python
rozsah = 5 # Tabulka 5x5

for radek in range(1, rozsah + 1):
    for sloupec in range(1, rozsah + 1):
        soucin = radek * sloupec
        # f"{soucin:2}" zajisti, ze cislo zabere vzdy 2 znaky (zarovnani)
        print(f"{soucin:2}", end="  ") 
    print() # Novy radek po dokonceni vnitrniho cyklu
````
