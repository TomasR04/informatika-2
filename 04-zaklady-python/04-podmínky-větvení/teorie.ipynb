# 04 - Podmínky a Větvení

Větvení umožňuje programu reagovat na různé situace. V Pythonu k tomu slouží klíčová slova `if`, `elif` a `else`.

## 1. Jednoduchá podmínka (if)
Spusťte buňku. Změňte hodnotu `teplota` na menší než 100 a spusťte znovu. Všimněte si, že se text nevypíše.

```python
teplota = 150

if teplota >= 100:
    print("Pozor! Voda se vaří!")
    print("Vypínám konvici.")

print("Aktuální teplota:", teplota)
```

## 2. Větvení (if - else)
Pokrýváme obě varianty - pravda i nepravda.

```python
heslo = "tajne"
zadano = input("Zadej heslo: ")

if zadano == heslo:
    print("Přístup povolen.")
else:
    print("Přístup zamítnut.")
```

## 3. Více možností (if - elif - else)
Hodnocení studenta podle bodů. Python prochází podmínky shora dolů. Jakmile jedna platí, ostatní ignoruje.

```python
body = int(input("Zadej počet bodů (0-100): "))

if body >= 90:
    print("Známka: 1 (Výborně)")
elif body >= 75:
    print("Známka: 2 (Chvalitebně)")
elif body >= 50:
    print("Známka: 3 (Dobře)")
else:
    print("Známka: 4 (Neprospěl)")
```

## 4. Ternární operátor
Velmi užitečné pro rychlé nastavení proměnné. Místo 4 řádků kódu stačí 1.

```python
vek = 17

# Dlouhý zápis:
if vek >= 18:
    status = "Dospělý"
else:
    status = "Dítě"
print(f"Status (dlouze): {status}")

# Zkrácený zápis (Ternární operátor):
status_short = "Dospělý" if vek >= 18 else "Dítě"
print(f"Status (krátce): {status_short}")
```

## 5. Match - Case (Switch)
Moderní konstrukce dostupná v novějších verzích Pythonu. Ideální pro reakci na konkrétní hodnoty.

```python
den = int(input("Zadej číslo dne v týdnu (1-7): "))

match den:
    case 1:
        print("Pondělí")
    case 2:
        print("Úterý")
    case 3:
        print("Středa")
    case 4:
        print("Čtvrtek")
    case 5:
        print("Pátek")
    case 6 | 7:  # Svislítko | znamená NEBO (or)
        print("Víkend!")
    case _:
        print("Neplatný den")
```
